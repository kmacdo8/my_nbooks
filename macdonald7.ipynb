```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('avocado.csv')
df.head()
```

```python
df.info()
```

```python
df[['Date']] = df[['Date']].apply(pd.to_datetime)
df.info()
```

```python
df_filtered = df.query('year == 2015' and 'type == "conventional"' and 
        'region == "Albany"')[['Date', 'Total Bags',
                               'Small Bags']]
df_filtered.head()
```

```python
df_melted = pd.melt(df_filtered, id_vars='Date',
                    value_vars=['Total Bags','Small Bags'],
                    var_name='Bags', value_name='Count')
df_melted
```

```python
sns.relplot(data=df_melted, kind='line',
            x='Date', y='Count',
            hue='Bags')
```

```python
df['Sales Volume'] = pd.qcut(df['Total Volume'], q=4,
                             labels=['Poor', 'Modest', 
                                     'Good', 'Excellent'])
df.head()
```

```python
sns.countplot(data=df,x='year', hue='Sales Volume')
```

```python
df_size_melt = pd.melt(df, id_vars='year',
                      value_vars=['Small Bags', 'Large Bags', 'XLarge Bags'],
                      var_name='Bags', value_name='Count')
df_size_melt
```

```python
sns.lineplot(data=df_size_melt,
            x='year', y='Count', hue='Bags')
```
